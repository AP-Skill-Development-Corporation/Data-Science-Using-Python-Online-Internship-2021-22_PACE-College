### Regression plot:
* Regression plot create a regression line between features
* it is used for find the linear relationship

```python
# Seaborn provides a method load_datasets for create dataframe
df=sns.load_dataset('tips')
df
```

```python
# how to find how many rows and columns
df.shape
```

```python
# find the information about dataset
df.info()
```

```python
# null values is present or not
df.isnull().sum()
```

```python
# It return the statstical information about dataset
df.describe()
```

```python
# find the correlation between features
df.corr()
```

```python
sns.lmplot(x='total_bill',y='tip',data=df)
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=df)
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=df,markers=['d','s'])
```

```python
help(sns.lmplot)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=df,markers=["o", "x"])
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=df,palette=dict(Yes='g',No='m'))
```

### joint plot

* It is combination of two plots
* Default plot is scatter and Hist
* Use the relation b/w to variables


```python
# iris dataset is a flower dataset
df=sns.load_dataset('iris')
df.sample(5)
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=df)
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=df,kind='reg')
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=df,kind='reg')
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=df,kind='kde')
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=df,kind='hex')
```

### Pairplot:
* Pairplot is used for display the multiple plots at a time
* It gives pair-wise relationship
* It is also used for find the feature selection

```python
sns.pairplot(df,hue='species')
```

### HeatMap:

* It is used for graphical reprasentation of correleation values
### correlation: 
* find the relation b/w two features
* it is also used for find the feature selection
* it denotes from -1 to 1
* if correlation value is -1 => it reprasents the negative relation
* if correlation value is 0 => No relation b/w features
* if correlation value is 1 => it reprasents the positive relation


```python
c=df.corr()
c
```

```python
sns.heatmap(c)
```

```python
sns.heatmap(c,annot=True)
```

```python
sns.heatmap(c,annot=True,linecolor='red',linewidth=1)
```

```python
sns.heatmap(c,annot=True,linecolor='red',linewidth=1,cmap="YlGnBu")
#help(sns.heatmap)
```
