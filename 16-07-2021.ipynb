```python
# how to add legend => short description
x=np.linspace(0,4,50)
plt.plot(x,x,'g',label="first")
plt.plot(x,x**3,'m',label="second")
plt.legend()
plt.show()
```

```python
x=np.arange(2,20,2)
plt.plot(x,x**2,'b--',label="dashed line")
plt.plot(x,x**3,'rs',label="squared")
plt.legend()
plt.show()
```

```python
x=np.arange(2,20,2)
plt.plot(x,x**2,'b--',label="dashed line")
plt.plot(x,x**2.2,'rs',label="squared")
plt.plot(x,x**2.5,'g^',label="traingle")
plt.legend()
plt.show()
```

```python
# change the legend position
#help(plt.legend)
x=np.arange(2,20,2)
plt.plot(x,x**2,'b--',label="dashed line")
plt.plot(x,x**2.2,'rs',label="squared")
plt.plot(x,x**2.5,'g^',label="traingle")
plt.legend(loc=9)    # loc=from 0 to 9
plt.show()
```

```python
# change the legend position in outside
#help(plt.legend)
x=np.arange(2,20,2)
plt.plot(x,x**2,'b--',label="dashed line")
plt.plot(x,x**2.2,'rs',label="squared")
plt.plot(x,x**2.5,'g^',label="traingle")
#plt.legend(loc=(1,0))   
plt.legend(loc=(1.1,0.3))    
plt.show()
```

```python
# change the legend position in outside
#help(plt.legend)

title_dict={'color':'green', 'size':19, 'style':'italic'}
x=np.arange(2,20,2)
plt.plot(x,x**2,'b--',label="dashed line")
plt.plot(x,x**2.2,'rs',label="squared")
plt.plot(x,x**2.5,'g^',label="traingle")
plt.title("Sample Graph",fontdict=title_dict)
plt.xlabel("x-axis",fontdict=title_dict)
plt.ylabel("y-axis",fontdict=title_dict)
#plt.legend(loc=(1,0))   
plt.legend(loc=(1.1,0.3))    
plt.show()
```

### Scatter plot

```python
x=[11,2,30,45,15]
y=[10,20,30,40,50]
plt.scatter(x,y)
plt.title("sample graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x=[11,2,30,45,15]
y=[10,20,30,40,50]
plt.scatter(x,y)
plt.title("sample graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

for a,b in zip(x,y):
    plt.text(a,b,(str(a),str(b)))

plt.show()
```

### subplots:

* we can divide the figure into multiple subplots
* syntax: plt.subplot(row,col,pos)

```python
x=np.array([10,20,30,40])
y=np.array([50,60,70,80])

plt.subplot(2,2,1)
plt.plot(x,y,'r')

plt.subplot(2,2,2)
plt.plot(x,y**3,'g')

plt.subplot(2,2,3)
plt.plot(x,x*y,'r')

plt.subplot(2,2,4)
plt.plot(x,y,'r--')
plt.show()
```

### Bar plots

```python
x=['jan','feb','mar','apr']
y=[250,340,200,550]
plt.bar(x,y)
plt.title("recovary rate in 2021")
plt.xlabel("Month")
plt.ylabel("recovary")
plt.show()
```

```python
x=['jan','feb','mar','apr']
y=[250,340,200,550]
plt.bar(x,y)
plt.title("recovary rate in 2021")
plt.xlabel("Month")
plt.ylabel("recovary")

for a,b in zip(x,y):
    plt.text(a,b,str(b))
plt.show()
```

```python
# Horizontal bar graph
x=['jan','feb','mar','apr']
y=[250,340,200,550]
plt.barh(x,y)
plt.title("recovary rate in 2021")
plt.xlabel("Month")
plt.ylabel("recovary")

for a,b in zip(x,y):
    plt.text(b,a,str(b))
plt.show()
```

### Hist plot

```python
x=np.random.randn(1000)
plt.hist(x,histtype='step')
plt.show()
```

### Box plot

* Box plot is used for find the min,max, 25%,75%, mean
* By using box plot we can find the outliers in dataset

```python
x=np.random.normal(100,20,200)
plt.boxplot(x)
plt.show()
```

### Pie plot

```python
course=['python','datascice','java','c']
size=[200,250,200,150]
plt.pie(size,labels=course)
plt.show()
```

```python
course=['python','datascice','java','c']
size=[200,250,200,150]
plt.pie(size,labels=course,autopct='%1.2f%%')
plt.show()
```

```python
course=['python','datascice','java','c']
size=[200,250,200,150]
plt.pie(size,labels=course,explode=[0.1,0,0,0],autopct='%1.2f%%')
plt.show()
```

```python
#save fig
course=['python','datascice','java','c']
size=[200,250,200,150]
plt.pie(size,labels=course,explode=[0.1,0,0,0],autopct='%1.2f%%')
plt.savefig('mychat.png')
plt.show()
```
