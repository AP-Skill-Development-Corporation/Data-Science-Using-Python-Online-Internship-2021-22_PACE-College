### Group

```python
url="https://raw.githubusercontent.com/AP-Skill-Development-Corporation/Data-Science-Using-Python-Online-Internship-2021-22_PACE-College/main/USA_cars_datasets.csv"
df=pd.read_csv(url)
df.head()
```

```python
# how to delete the column
df=df.drop('Unnamed: 0',axis=1)
df
```

```python
df['country'].unique()
```

```python
df['country'].value_counts()
```

```python
g=df.groupby('country')
g
```

```python
for i, j in g:
    print(i,j)
```
