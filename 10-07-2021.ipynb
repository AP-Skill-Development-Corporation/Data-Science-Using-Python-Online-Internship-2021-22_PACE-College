### How to access the data using files
* CSV    => pd.read_csv('file name')
* excel  => pd.read_excel('file name')
* Json   => pd.read_json('json file')
* Html   => pd.read_html(url)

### Convert your dataframe into files
* ex: pd.to_csv/to_excel/to_json/to_html

```python
# Read csv file
df=pd.read_csv('studentdata.csv')    # path: 'D:/foldername/filename.csv'
df
```

```python
# convert csv file into excel
df.to_excel('studenexcel.xlsx')
```

```python
# read excel into dataframe
df2=pd.read_excel('studenexcel.xlsx')
df2
```

```python
# convert csv file into json
# json: java script object notation
# json object({}); json array([]); json string(key:value)

json_data=df.to_json(orient='records')
json_data
```

```python
# convert json file into dataframe
df3=pd.read_json(json_data)
df3
```

```python
# read html data into dataframe
import requests
url="https://www.worldcoinindex.com/"
res=requests.get(url)
res
```

```python
df=pd.read_html(res.text)
df
df=df[0]
df=df.iloc[:,2:]
df
```
